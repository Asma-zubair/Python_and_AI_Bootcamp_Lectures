# **Sets**

A set in Python is a **collection of unique items**.

It is mostly used when you want to remove duplicates  or check membership quickly.

**✔ Characteristics of Sets:**

*   Written in curly braces → { }

*   No duplicates allowed
*   Unordered (items have no index)


*   Changeable (you can add/remove items)




```python
numbers = {1, 2, 3, 4}
```

```python
duplicates = {1, 2, 2, 3}
print(duplicates)
# Output: {1, 2, 3}
# Python automatically removes duplicate "2"
```

**Set Operations**

**1. Creating a Set**

```python
my_set = {10, 20, 30}
print(my_set)
```

**2. Adding Items**

```python
fruits = {"apple", "banana"}
fruits.add("mango")
print(fruits)
```

 **3. Removing Items**

```python
fruits.remove("banana")
print(fruits)
```

**4. Checking If Value Exists**

```python
numbers = {5, 10, 15, 20}
print(10 in numbers)   # True
print(50 in numbers)   # False
```

**5. Set Length**

```python
print(len(numbers))
```

**6. Removing Duplicates Using Set**

```python
data = [1, 2, 2, 3, 3, 3, 4]
unique_data = set(data)
print(unique_data)  # {1, 2, 3, 4}
```

**7. Union (Combine Sets)**

```python
a = {1, 2, 3}
b = {3, 4, 5}

print(a.union(b))
# Output: {1, 2, 3, 4, 5}
```

**8. Intersection (Common Values)**

```python
a = {1, 2, 3}
b = {2, 3, 4}

print(a.intersection(b))
# Output: {2, 3}
```
